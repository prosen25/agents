```python
from dotenv import load_dotenv
from agents.mcp import MCPServerStdio
from agents import Agent, trace, Runner
from IPython.display import display, Markdown

load_dotenv(override=True)
```

```python
params = {
    "command": "/home/prosen/.nvm/versions/node/v24.14.1/bin/npx",
    "args": ["-y", "mcp-memory-libsql"],
    "env": {"LIBSQL_URL": "file:./memory/prosen.db"}
}

async with MCPServerStdio(params=params, client_session_timeout_seconds=30) as server:
    mcp_tools = await server.list_tools()
```

```python
mcp_tools
```

```python
instructions = """You use your entity tools as a persistent memory to store 
and recall information about your conversations."""
request = """My name's Prosen. I'm an LLM engineer. I'm teaching a course about AI Agents, 
including the incredible MCP protocol. MCP is a protocol for connecting agents with tools, 
resources and prompt templates, and makes it easy to integrate AI agents with capabilities."""
model = "gpt-4.1-mini"
```

```python
async with MCPServerStdio(params=params, client_session_timeout_seconds=30) as mcp_server:
    agent = Agent(name="conversation_agent", instructions=instructions, model=model, mcp_servers=[mcp_server])
    with trace("conversation"):
        result = await Runner.run(starting_agent=agent, input=request)

    display(Markdown(result.final_output))
```
